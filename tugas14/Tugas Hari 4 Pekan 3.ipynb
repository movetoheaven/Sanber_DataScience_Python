<h2>Soal 4: Perbandingan Kuantitatif Barplot - Grouping Visualisasi Dengan Barplot</h2>

1. Visualisasikan gambar yang tadi kita manipulasi, untuk menghasilkan visualisasi seperti di bawah ini
2. Tuliskan apa insight yang bisa kamu dapat dari visualisasi tersebut

```python
import matplotlib.pyplot as plt

# Group the data and calculate the mean
mean_sales_by_genre = df.groupby('Genre')[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']].mean()

# Extract genre names
genres = mean_sales_by_genre.index

# Extract mean sales data for each region
na_sales = mean_sales_by_genre['NA_Sales']
eu_sales = mean_sales_by_genre['EU_Sales']
jp_sales = mean_sales_by_genre['JP_Sales']
other_sales = mean_sales_by_genre['Other_Sales']

# Set the width of the bars
bar_width = 0.2

# Create a figure and axis
fig, ax = plt.subplots(figsize=(12, 6))

# Create bar positions for each group
bar_positions = range(len(genres))
bar_positions_na = [pos - bar_width for pos in bar_positions]
bar_positions_eu = bar_positions
bar_positions_jp = [pos + bar_width for pos in bar_positions]
bar_positions_other = [pos + 2 * bar_width for pos in bar_positions]

# Create the bar plots for each region
ax.bar(bar_positions_na, na_sales, width=bar_width, label='NA_Sales', align='center')
ax.bar(bar_positions_eu, eu_sales, width=bar_width, label='EU_Sales', align='center')
ax.bar(bar_positions_jp, jp_sales, width=bar_width, label='JP_Sales', align='center')
ax.bar(bar_positions_other, other_sales, width=bar_width, label='Other_Sales', align='center')

# Set labels for x-axis ticks
ax.set_xticks(bar_positions)
ax.set_xticklabels(genres, rotation=90)

# Set labels for y-axis and title
ax.set_ylabel('Mean Sales')
ax.set_title('Mean Sales by Genre and Region')

# Add legend
ax.legend(title='Region')

# Display the plot
plt.show()
```
