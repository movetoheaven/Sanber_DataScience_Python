<h2>Soal 1: Multiple Subplots</h2>

Dengan data di bawah ini buatlah visualisasi seperti expected output:

```python

import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(2*-np.pi, 2*np.pi, 200)

tan = np.tan(x)/10
cos = np.cos(x)
sin = np.sin(x)
```

```python
#code here
rows = 1; cols = 3;

fig, ax = plt.subplots(rows, cols, figsize=(20, 8), sharex= True, sharey=True)
ax[0].plot(x, tan, color='b', label='tanX')
ax[1].plot(x, sin, color='r', label='sinX')
ax[2].plot(x, cos, color='g', label='cosX')

ax[0].set_title('Grafik tanX')
ax[1].set_title('Grafik sinX')
ax[2].set_title('Grafik cosX')

fig.legend(loc='upper center', ncols=1)

plt.show()
```

<h2>Soal 2: Nested Axis</h2>

Dengan data di bawah ini, buatlah visualisasi seperti expected output:

```python
x = np.linspace(2*-np.pi, 2*np.pi, 100)
y = np.cos(x)
y2 = np.cos(x**2)
y3 = np.cos(x**3)
y4 = np.cos(x**4)
y5 = np.cos(x**5)
```

```python
#code here
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.axes_grid1.inset_locator import inset_axes

# Generate the x values
x = np.linspace(2 * -np.pi, 2 * np.pi, 100)

# Create the main figure
fig, ax = plt.subplots(figsize=(10, 6))
fig.suptitle('Grafik CosX')

# Plot cos(x) on the main axes
ax.plot(x, np.cos(x), label='cos(x)')

# Create inset axes for cos(x^2)
axins1 = inset_axes(ax, width='30%', height='30%', loc='upper right')
axins1.plot(x, y2, label='cos(x^2)', color='red')
axins1.set_xticks([]), axins1.set_yticks([])

# Create inset axes for cos(x^3)
axins2 = inset_axes(ax, width='30%', height='30%', loc='upper left')
axins2.plot(x, y3, label='cos(x^3)', color='green')
axins2.set_xticks([]), axins2.set_yticks([])

# Create inset axes for cos(x^4)
axins3 = inset_axes(ax, width='30%', height='30%', loc='lower left')
axins3.plot(x, y4, label='cos(x^4)', color='black')
axins3.set_xticks([]), axins3.set_yticks([])

# Create inset axes for cos(x^5)
axins4 = inset_axes(ax, width='30%', height='30%', loc='lower right')
axins4.plot(x, y5, label='cos(x^5)',color='yellow')
axins4.set_xticks([]), axins4.set_yticks([])

# Display a legend
ax.legend(loc='upper right')

plt.show()
```
